#### Rescan duplicate list then begin to merge

```python

merge_dict = {}

merge_path = "D:/final_merge/"
# scan
merge_list = os.scandir(merge_path)

for pdf in merge_list:
    #print(pdf.path[:25])
    if pdf.path[:25] in merge_dict.keys():
        temp = merge_dict[pdf.path[:25]]
        merge_dict[pdf.path[:25]] = temp + 1
    elif pdf.path[:25] not in merge_dict.keys():
        merge_dict[pdf.path[:25]] = 1
count = 0
print(len(merge_dict))
for key, value in merge_dict.items():    
    print(key,value)
    count += 1 * value
print(count)
#print("batch 1 size:", len(pdf_list_1), " pdfs")
```

```python
# merge
import PyPDF2
from PyPDF2 import PdfReader , PdfWriter, PdfMerger

corrupted = []
destination = ""
count = 0
#
for key, value in merge_dict.items():
    merger = PdfWriter()
    base_pdf = key + ".pdf"
    #print("base pdf=", base_pdf)
    merger.append(base_pdf)
    # add second pdf
    second_pdf = key + "2.pdf"
    #print("second pdf=", second_pdf)
    merger.append(second_pdf)
    # add third is third pdf exist
    if value == 3:
        third_pdf = key + "3.pdf"
        #print("third pdf=", third_pdf)
        merger.append(third_pdf)
    # finished merging pdfs so save with tin as new pdf name
    try:
        destination = "D:/" + key[15:25] + ".pdf"
        merger.write(destination)
        print("merged-", key[15:25], "to", destination)
        count += 1
    except Exception:
        print("error saving ", destination)
        corrupted.append(key)
    finally:
        merger.close()
print("Merged {} pdfs".format(count))
```
