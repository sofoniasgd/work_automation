```python
import os
from pathlib import Path
file_root_dir = "D:/efiling"
print(os.path.exists(file_root_dir))
```

```python
entries = os.scandir(file_root_dir)
list = []
move_list = []


for entry in entries:
    # find pdfs less than ten digit in number
    if entry.is_file():
        # check if current file name length is more than tendigits(14 with extention)
        # and check if its base file exists(if its a additional pdf of another)
        base_file = "D:/efiling/" + entry.name[:10] + '.pdf'
        #print(type(base_file))
        # file is a part of a base file so record tin
        if len(entry.name) > 14 and os.path.isfile(base_file):
            #print(entry.path[:21])
            move_list.append(entry.path)
            if entry.path[:21] not in list:
                list.append(entry.path[:21])


print(len(move_list), " total pdfs will be merged ", end='')
print("into ", len(list) , " base pdfs\n")
print("base pdfs:")
for line in list:
    print(line)
print("mergeable  pdfs:")
for line in list:
    print(line)
# move all pdfs to separate dir so merging will be easier
```

```python
import shutil
dest = "D:/"
print(os.path.exists(dest))
```

```python
#MOVE THE MERGEABLE FILES
moved = 0
for path in move_list:
    if os.path.isfile(path):
        shutil.move(path, dest)
        moved += 1
print("moved ", moved, "pdfs")
```

```python
# MOVE BASE PDFS
moved = 0
for path1 in list:
    print("checking", path1)
    path = path1 + ".pdf"
    if os.path.exists(path):
        shutil.move(path, dest)
        print("found file, moving")
        moved += 1
print("moved ", moved, "pdfs")
```

```python
# create new list for base pdfs with full path
base_list = []
merge_list = []

entries = os.scandir("D:/")

for entry in entries:
    if os.path.isfile(entry.path) and len(entry.name) == 14:
        base_list.append(entry.path)
    elif os.path.isfile(entry.path) and len(entry.name) != 14:
        merge_list.append(entry.path)
print("base list= ", len(base_list))
print(base_list)
print("merge list= ", len(merge_list))
print(merge_list)
```

```python
import PyPDF2
from PyPDF2 import PdfReader , PdfWriter, PdfMerger
merged_list = []
corrupted = []
count = 0
#
for base_pdf in base_list:
    # for every base file, look for its parts in merge_list then merge
    merger = PdfWriter()
    if os.path.isfile(base_pdf):
        merger.append(base_pdf)
        # find and append all part pdfs
        for part_pdf in merge_list:
            if (part_pdf[3:13] == base_pdf[3:13]):
                merger.append(part_pdf)
        
        # finished merging pdfs so save with tin as new pdf name
        try:
            merger.write(base_pdf[:13] + "x.pdf")
            print("merged-", base_pdf[3:13])
            count += 1
        except Exception:
            print("error saving ", base_pdf)
            corrupted.append(base_pdf)
        finally:
            merger.close()
print("successfully merged= ", count)
print("error=", corrupted)
```
